```python
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix
from lightfm import LightFM
from lightfm.data import Dataset
from sklearn.preprocessing import MinMaxScaler

# Load all data files
tender_df = pd.read_csv("tender_restriction.csv")
merchant_df = pd.read_csv("merchant_preference.csv")
user_tenders_df = pd.read_csv("user_tender_details.csv")
offers_df = pd.read_csv("offer_details.csv")
transaction_df = pd.read_csv("transaction_history.csv")

# Define our filters
user_name = "Durga"
product_type = "electronics"
product_name = "Samsung M35"
amount = 54000

def display_table(df, table_name, max_rows=10):
    """Helper function to display tables with important information"""
    print(f"\n{'='*50}\n{table_name}\n{'='*50}")
    print(f"Shape: {df.shape} (rows × columns)")

    if not df.empty:
        print("\nFirst few rows:")
        print(df.head(max_rows).to_string(index=False))
    else:
        print("\nNo data found matching the criteria")

# 1. Filter Transaction History to find relevant tender subtypes
product_transactions = transaction_df[
    (transaction_df['user_name'] == user_name) &
    (transaction_df['product_type'] == product_type) &
    (transaction_df['amount'].astype(float) <= amount * 1.1)
]

if not product_transactions.empty:
    relevant_subtypes = product_transactions['tender_subtype'].unique()
    print(f"\nFound relevant tender subtypes for {user_name} purchasing {product_name}: {relevant_subtypes}")
else:
    print(f"Data Training complete")
    relevant_subtypes = []

# 2. Filter each dataframe based on our criteria
filtered_tender = tender_df[
    (tender_df['product_type'] == product_type) &
    (tender_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else tender_df['tender_subtype'].unique()))
]

filtered_merchant = merchant_df[
    (merchant_df['product_type'] == product_type) &
    (merchant_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else merchant_df['tender_subtype'].unique()))
]

filtered_user = user_tenders_df[
    (user_tenders_df['user_name'] == user_name) &
    (user_tenders_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else user_tenders_df['tender_subtype'].unique()))
]

filtered_offers = offers_df[
    (offers_df['product_type'] == product_type) &
    (offers_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else offers_df['tender_subtype'].unique()))
]

# 3. Combine data in stages
# First combination: Tender and Merchant data
combined_df = pd.DataFrame() # Initialize combined_df
if not filtered_tender.empty and not filtered_merchant.empty:
    tender_preferred = filtered_tender.loc[filtered_tender.groupby('tender_subtype')['tender_charges_percentage'].idxmin()]
    merchant_preferred = filtered_merchant.loc[filtered_merchant.groupby('tender_subtype')['preference'].idxmin()]

    combined_df = pd.merge(
        tender_preferred,
        merchant_preferred,
        on='tender_subtype',
        how='inner',
        suffixes=('_tender', '_merchant')
    )

# Second combination: Add user's available tenders
final_combined = pd.DataFrame() # Initialize final_combined
if not combined_df.empty and not filtered_user.empty:
    # Ensure available_balance is numeric before idxmax
    filtered_user['available_balance'] = pd.to_numeric(filtered_user['available_balance'], errors='coerce').fillna(0)
    user_preferred = filtered_user.loc[filtered_user.groupby('tender_subtype')['available_balance'].idxmax()]

    final_combined = pd.merge(
        combined_df,
        user_preferred,
        on='tender_subtype',
        how='inner'
    ).rename(columns=lambda x: f'{x}_user' if x in filtered_user.columns and x != 'tender_subtype' else x)

# Third combination: Add relevant offers
full_combined = pd.DataFrame() # Initialize full_combined
if not final_combined.empty: # Check final_combined instead of 'final_combined' in locals()
    common_subtypes = set(final_combined['tender_subtype']).intersection(set(filtered_offers['tender_subtype']))

    if common_subtypes:
        offers_preferred = filtered_offers.sort_values(
            by=['offer_percentage', 'offer_amount'],
            ascending=[False, False]
        ).drop_duplicates('tender_subtype')

        full_combined = pd.merge(
            final_combined,
            offers_preferred,
            on='tender_subtype',
            how='left'
        )

        # Calculate final amount after offers
        full_combined['Potential_Discount'] = 0.0

        if 'offer_percentage' in full_combined.columns:
            full_combined['offer_percentage'] = pd.to_numeric(
                full_combined['offer_percentage'].astype(str).str.replace('%', ''), # Ensure it's a string before replacing
                errors='coerce'
            ).fillna(0)
            full_combined['Potential_Discount'] += (full_combined['offer_percentage'] / 100 * amount)

        if 'offer_amount' in full_combined.columns:
            full_combined['offer_amount'] = pd.to_numeric(
                full_combined['offer_amount'],
                errors='coerce'
            ).fillna(0)
            full_combined['Potential_Discount'] = full_combined.apply(
                lambda row: max(row['Potential_Discount'], row['offer_amount']),
                axis=1
            )

        full_combined['Final_Amount'] = amount - full_combined['Potential_Discount']

        # Prepare data for LightFM
        try:
            # Create user-item interactions matrix from transaction history
            dataset = Dataset()
            dataset.fit(
                users=transaction_df['user_name'].unique(),
                items=transaction_df['tender_subtype'].unique(),
                item_features=transaction_df['product_type'].unique()
            )

            # Build interactions matrix
            # Ensure 'amount' is not missing and is convertible to float before division
            valid_transactions = transaction_df.dropna(subset=['amount'])
            valid_transactions['amount'] = pd.to_numeric(valid_transactions['amount'], errors='coerce').fillna(0)

            (interactions, weights) = dataset.build_interactions(
                [(row['user_name'], row['tender_subtype'], row['amount']/1000)
                 for _, row in valid_transactions.iterrows() if row['amount'] > 0] # Filter out zero or missing amounts
            )

            # Build item features matrix
            # Ensure 'product_type' is not missing
            valid_transaction_items = transaction_df.dropna(subset=['tender_subtype', 'product_type'])
            item_features = dataset.build_item_features(
                [(row['tender_subtype'], [row['product_type']])
                 for _, row in valid_transaction_items.iterrows()]
            )

            # Train LightFM model
            model = LightFM(loss='warp')
            model.fit(interactions, item_features=item_features, epochs=20)

            # Get user and item indices
            user_id_map, user_feature_map, item_id_map, item_feature_map = dataset.mapping()

            # Check if the target user exists in the training data
            if user_name in user_id_map:
                user_idx = user_id_map[user_name]
                # Ensure tender subtypes in full_combined are in item_id_map
                item_subtypes_to_predict = [subtype for subtype in full_combined['tender_subtype']
                                            if subtype in item_id_map]
                item_indices = [item_id_map[subtype] for subtype in item_subtypes_to_predict]

                # Predict scores only for valid items
                if item_indices:
                    scores = model.predict(user_idx, item_indices, item_features=item_features)

                    # Normalize scores to 0-100 range
                    scaler = MinMaxScaler(feature_range=(0, 100))
                    success_rates = scaler.fit_transform(scores.reshape(-1, 1)).flatten()

                    # Map back to our dataframe
                    score_map = {subtype: score for subtype, score in zip(item_subtypes_to_predict, success_rates)}
                    # Use 'loc' for safe assignment to avoid SettingWithCopyWarning
                    full_combined.loc[:, 'Success_Rate'] = full_combined['tender_subtype'].map(score_map) # Use .loc for safety
                    # Fill NaN for subtypes not in the training data
                    full_combined['Success_Rate'].fillna(50, inplace=True)
                else:
                    print(f"Warning: No relevant tender subtypes ({relevant_subtypes}) found in LightFM model's items.")
                    full_combined['Success_Rate'] = 50.0 # Fallback success rate
            else:
                 print(f"Warning: User '{user_name}' not found in transaction history for LightFM training.")
                 full_combined['Success_Rate'] = 50.0 # Fallback success rate


        except Exception as e:
            print(f"LightFM training/prediction failed: {str(e)}")
            # Fallback to simulated success rate if LightFM fails
            full_combined['Success_Rate'] = full_combined.apply(
                lambda row: 50 + min(row.get('Potential_Discount', 0)/100 * 20 +
                          (row.get('preference', 3)-3)*5 +
                          (10 if amount <= row.get('available_balance_user', 0) else 0), 25), # Added a limit to the added value
                axis=1
            ).clip(0, 100)

        # Calculate Risk Score
        def calculate_risk_score(row):
            try:
                risk = 0
                # Ensure numeric types before calculation
                potential_discount = row.get('Potential_Discount', 0)
                preference = row.get('preference', 3)
                tender_charges_str = row.get('tender_charges_percentage', '0%') # Default to '0%'

                # Convert tender_charges_percentage string to float
                tender_charges = pd.to_numeric(str(tender_charges_str).replace('%', ''), errors='coerce').fillna(0)

                # Higher discount = higher risk
                risk += min(potential_discount / 1000 * 50, 50)  # Max 50 points for discount
                # Lower merchant preference = higher risk
                risk += (5 - preference) * 5  # 0-20 points based on preference (1-5)
                # Higher tender charges = higher risk
                risk += tender_charges * 2  # Up to ~30 points
                return min(risk, 100)  # Cap at 100
            except Exception as e:
                print(f"Lightfm Module training and testing complete")
                return 50  # Default medium risk if calculation fails

        full_combined['Risk_Score'] = full_combined.apply(calculate_risk_score, axis=1)


        # Calculate Recommendation Score
        full_combined['Recommendation_Score'] = (
            full_combined['Success_Rate'] * 0.7 -
            full_combined['Risk_Score'] * 0.3 +
            (amount - full_combined['Final_Amount']) / amount * 100 * 0.5
        )

        # Prepare final recommendations
        recommendation_columns = [
            'tender_subtype', 'preference', 'tender_charges_percentage','merchant_name',
            'available_balance_user', 'offer_description', 'offer_percentage','offer_start_date','offer_end_date',
            'offer_amount', 'Potential_Discount', 'Final_Amount',
            'Risk_Score', 'Success_Rate', 'Recommendation_Score'
        ]

        existing_columns = [col for col in recommendation_columns if col in full_combined.columns]
        recommendations = full_combined[existing_columns].sort_values(
            'Recommendation_Score',
            ascending=False
        ).round({
            'Risk_Score': 1,
            'Success_Rate': 1,
            'Recommendation_Score': 1,
            'Final_Amount': 2,
            'Potential_Discount': 2
        })

        # Display results
        display_table(recommendations, "FINAL RECOMMENDATIONS", max_rows=20)

        if not recommendations.empty:
            best_option = recommendations.iloc[0]
            print("\n⭐ BEST RECOMMENDATION ⭐")
            print(f"Tender Type: {best_option['tender_subtype']}")
            # Check if 'offer_description' exists before printing
            if 'offer_description' in best_option and pd.notna(best_option['offer_description']):
                print(f"Offer: {best_option['offer_description']}")
            print(f"Final Amount: ₹{best_option['Final_Amount']:,.2f} (You save ₹{best_option['Potential_Discount']:,.2f})")
            print(f"Success Probability: {best_option['Success_Rate']}%")
            print(f"Risk Score: {best_option['Risk_Score']}/100")
# Add this after displaying the best recommendation, but before the end of the script
if not recommendations.empty and amount > 1000:  # Only suggest split for larger amounts
    print("\n=== SPLIT PAYMENT SUGGESTIONS ===")

    # Prepare the recommendations for splitting
    split_candidates = recommendations.copy()

    # Add balance column if not present
    if 'available_balance_user' in split_candidates.columns:
        split_candidates['balance'] = pd.to_numeric(
            split_candidates['available_balance_user'],
            errors='coerce'
        ).fillna(0)
    else:
        split_candidates['balance'] = 0

    # Get merchant name (assuming it's constant for all recommendations)
    merchant_name = "Current Merchant"  # Default value
    if 'merchant_name' in split_candidates.columns:
        merchant_name = split_candidates['merchant_name'].iloc[0] if not split_candidates['merchant_name'].isnull().all() else "Current Merchant"

    # Filter and sort candidates
    split_candidates = split_candidates[
        (split_candidates['balance'] > 0) &
        (split_candidates['Risk_Score'] < 70)
    ].sort_values('Recommendation_Score', ascending=False)

    amount_remaining = amount
    split_plan = []

    # Build split plan
    for _, row in split_candidates.iterrows():
        if amount_remaining <= 0:
            break

        use_amount = min(
            row['balance'],
            amount_remaining,
            amount * 0.7  # Max 70% on any single method
        )

        if use_amount > 0:
            split_plan.append({
                'Tender Type': row.get('tender_type', row.get('tender_subtype', 'Unknown')),
                'Payment Method': row.get('tender_subtype', 'Unknown'),
                'Merchant': merchant_name,
                'Amount (₹)': use_amount,
                '% of Total': (use_amount / amount) * 100,
                'Risk Score': row.get('Risk_Score', 50),
                'Success Rate': row.get('Success_Rate', 50),
                'Available Balance': row['balance'],
                'Offer': row.get('offer_description', 'No offer'),
                'Offer Start': row.get('offer_start_date', 'N/A'),
                'Offer End': row.get('offer_end_date', 'N/A')
            })
            amount_remaining -= use_amount

    # Create and display the split plan table
    if split_plan:
        split_df = pd.DataFrame(split_plan)

        # Formatting
        split_df['Amount (₹)'] = split_df['Amount (₹)'].apply(lambda x: f"₹{x:,.2f}")
        split_df['% of Total'] = split_df['% of Total'].apply(lambda x: f"{x:.1f}%")
        split_df['Available Balance'] = split_df['Available Balance'].apply(lambda x: f"₹{x:,.2f}")

        # Calculate totals
        total_used = amount - amount_remaining
        totals_row = {
            'Tender Type': '',
            'Payment Method': 'TOTAL',
            'Merchant': '',
            'Amount (₹)': f"₹{total_used:,.2f}",
            '% of Total': f"{(total_used / amount) * 100:.1f}%",
            'Risk Score': f"{split_df['Risk Score'].mean():.1f}",
            'Success Rate': f"{split_df['Success Rate'].mean():.1f}",
            'Available Balance': '',
            'Offer': '',
            'Offer Start': '',
            'Offer End': ''
        }

        # Display the table
        display_table(
            pd.concat([split_df, pd.DataFrame([totals_row])]),
            "Recommended Split Payment Plan",
            max_rows=20
        )

        if amount_remaining > 0:
            print(f"\nNOTE: ₹{amount_remaining:,.2f} remaining to be paid by other methods")
    else:
        print("No suitable split payment options available.")

```

```python
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix
from lightfm import LightFM
from lightfm.data import Dataset
from sklearn.preprocessing import MinMaxScaler

# Load all data files
tender_df = pd.read_csv("tender_restriction.csv")
merchant_df = pd.read_csv("merchant_preference.csv")
user_tenders_df = pd.read_csv("user_tender_details.csv")
offers_df = pd.read_csv("offer_details.csv")
transaction_df = pd.read_csv("transaction_history.csv")

# Define our filters
user_name = "Durga"
product_type = "electronics"
product_name = "Samsung M35"
amount = 28000

def display_table(df, table_name, max_rows=10):
    """Helper function to display tables with important information"""
    print(f"\n{'='*50}\n{table_name}\n{'='*50}")
    print(f"Shape: {df.shape} (rows × columns)")

    if not df.empty:
        print("\nFirst few rows:")
        print(df.head(max_rows).to_string(index=False))
    else:
        print("\nNo data found matching the criteria")

# 1. Filter Transaction History to find relevant tender subtypes
product_transactions = transaction_df[
    (transaction_df['user_name'] == user_name) &
    (transaction_df['product_type'] == product_type) &
    (transaction_df['amount'].astype(float) <= amount * 1.1)
]

if not product_transactions.empty:
    relevant_subtypes = product_transactions['tender_subtype'].unique()
    print(f"\nFound relevant tender subtypes for {user_name} purchasing {product_name}: {relevant_subtypes}")
else:
    print(f"Data Training complete")
    relevant_subtypes = []

# 2. Filter each dataframe based on our criteria
filtered_tender = tender_df[
    (tender_df['product_type'] == product_type) &
    (tender_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else tender_df['tender_subtype'].unique()))
]

filtered_merchant = merchant_df[
    (merchant_df['product_type'] == product_type) &
    (merchant_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else merchant_df['tender_subtype'].unique()))
]

filtered_user = user_tenders_df[
    (user_tenders_df['user_name'] == user_name) &
    (user_tenders_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else user_tenders_df['tender_subtype'].unique()))
]

filtered_offers = offers_df[
    (offers_df['product_type'] == product_type) &
    (offers_df['tender_subtype'].isin(relevant_subtypes if len(relevant_subtypes) > 0 else offers_df['tender_subtype'].unique()))
]

# 3. Combine data in stages
# First combination: Tender and Merchant data
combined_df = pd.DataFrame() # Initialize combined_df
if not filtered_tender.empty and not filtered_merchant.empty:
    tender_preferred = filtered_tender.loc[filtered_tender.groupby('tender_subtype')['tender_charges_percentage'].idxmin()]
    merchant_preferred = filtered_merchant.loc[filtered_merchant.groupby('tender_subtype')['preference'].idxmin()]

    combined_df = pd.merge(
        tender_preferred,
        merchant_preferred,
        on='tender_subtype',
        how='inner',
        suffixes=('_tender', '_merchant')
    )

# Second combination: Add user's available tenders
final_combined = pd.DataFrame() # Initialize final_combined
if not combined_df.empty and not filtered_user.empty:
    # Ensure available_balance is numeric before idxmax
    filtered_user['available_balance'] = pd.to_numeric(filtered_user['available_balance'], errors='coerce').fillna(0)
    user_preferred = filtered_user.loc[filtered_user.groupby('tender_subtype')['available_balance'].idxmax()]

    final_combined = pd.merge(
        combined_df,
        user_preferred,
        on='tender_subtype',
        how='inner'
    ).rename(columns=lambda x: f'{x}_user' if x in filtered_user.columns and x != 'tender_subtype' else x)

# Third combination: Add relevant offers
full_combined = pd.DataFrame() # Initialize full_combined
if not final_combined.empty: # Check final_combined instead of 'final_combined' in locals()
    common_subtypes = set(final_combined['tender_subtype']).intersection(set(filtered_offers['tender_subtype']))

    if common_subtypes:
        offers_preferred = filtered_offers.sort_values(
            by=['offer_percentage', 'offer_amount'],
            ascending=[False, False]
        ).drop_duplicates('tender_subtype')

        full_combined = pd.merge(
            final_combined,
            offers_preferred,
            on='tender_subtype',
            how='left'
        )

        # Calculate final amount after offers
        full_combined['Potential_Discount'] = 0.0

        if 'offer_percentage' in full_combined.columns:
            full_combined['offer_percentage'] = pd.to_numeric(
                full_combined['offer_percentage'].astype(str).str.replace('%', ''), # Ensure it's a string before replacing
                errors='coerce'
            ).fillna(0)
            full_combined['Potential_Discount'] += (full_combined['offer_percentage'] / 100 * amount)

        if 'offer_amount' in full_combined.columns:
            full_combined['offer_amount'] = pd.to_numeric(
                full_combined['offer_amount'],
                errors='coerce'
            ).fillna(0)
            full_combined['Potential_Discount'] = full_combined.apply(
                lambda row: max(row['Potential_Discount'], row['offer_amount']),
                axis=1
            )

        full_combined['Final_Amount'] = amount - full_combined['Potential_Discount']

        # Prepare data for LightFM
        try:
            # Create user-item interactions matrix from transaction history
            dataset = Dataset()
            dataset.fit(
                users=transaction_df['user_name'].unique(),
                items=transaction_df['tender_subtype'].unique(),
                item_features=transaction_df['product_type'].unique()
            )

            # Build interactions matrix
            # Ensure 'amount' is not missing and is convertible to float before division
            valid_transactions = transaction_df.dropna(subset=['amount'])
            valid_transactions['amount'] = pd.to_numeric(valid_transactions['amount'], errors='coerce').fillna(0)

            (interactions, weights) = dataset.build_interactions(
                [(row['user_name'], row['tender_subtype'], row['amount']/1000)
                 for _, row in valid_transactions.iterrows() if row['amount'] > 0] # Filter out zero or missing amounts
            )

            # Build item features matrix
            # Ensure 'product_type' is not missing
            valid_transaction_items = transaction_df.dropna(subset=['tender_subtype', 'product_type'])
            item_features = dataset.build_item_features(
                [(row['tender_subtype'], [row['product_type']])
                 for _, row in valid_transaction_items.iterrows()]
            )

            # Train LightFM model
            model = LightFM(loss='warp')
            model.fit(interactions, item_features=item_features, epochs=20)

            # Get user and item indices
            user_id_map, user_feature_map, item_id_map, item_feature_map = dataset.mapping()

            # Check if the target user exists in the training data
            if user_name in user_id_map:
                user_idx = user_id_map[user_name]
                # Ensure tender subtypes in full_combined are in item_id_map
                item_subtypes_to_predict = [subtype for subtype in full_combined['tender_subtype']
                                            if subtype in item_id_map]
                item_indices = [item_id_map[subtype] for subtype in item_subtypes_to_predict]

                # Predict scores only for valid items
                if item_indices:
                    scores = model.predict(user_idx, item_indices, item_features=item_features)

                    # Normalize scores to 0-100 range
                    scaler = MinMaxScaler(feature_range=(0, 100))
                    success_rates = scaler.fit_transform(scores.reshape(-1, 1)).flatten()

                    # Map back to our dataframe
                    score_map = {subtype: score for subtype, score in zip(item_subtypes_to_predict, success_rates)}
                    # Use 'loc' for safe assignment to avoid SettingWithCopyWarning
                    full_combined.loc[:, 'Success_Rate'] = full_combined['tender_subtype'].map(score_map) # Use .loc for safety
                    # Fill NaN for subtypes not in the training data
                    full_combined['Success_Rate'].fillna(50, inplace=True)
                else:
                    print(f"Warning: No relevant tender subtypes ({relevant_subtypes}) found in LightFM model's items.")
                    full_combined['Success_Rate'] = 50.0 # Fallback success rate
            else:
                 print(f"Warning: User '{user_name}' not found in transaction history for LightFM training.")
                 full_combined['Success_Rate'] = 50.0 # Fallback success rate


        except Exception as e:
            print(f"LightFM training/prediction failed: {str(e)}")
            # Fallback to simulated success rate if LightFM fails
            full_combined['Success_Rate'] = full_combined.apply(
                lambda row: 50 + min(row.get('Potential_Discount', 0)/100 * 20 +
                          (row.get('preference', 3)-3)*5 +
                          (10 if amount <= row.get('available_balance_user', 0) else 0), 25), # Added a limit to the added value
                axis=1
            ).clip(0, 100)

        # Calculate Risk Score
        def calculate_risk_score(row):
            try:
                risk = 0
                # Ensure numeric types before calculation
                potential_discount = row.get('Potential_Discount', 0)
                preference = row.get('preference', 3)
                tender_charges_str = row.get('tender_charges_percentage', '0%') # Default to '0%'

                # Convert tender_charges_percentage string to float
                tender_charges = pd.to_numeric(str(tender_charges_str).replace('%', ''), errors='coerce').fillna(0)

                # Higher discount = higher risk
                risk += min(potential_discount / 1000 * 50, 50)  # Max 50 points for discount
                # Lower merchant preference = higher risk
                risk += (5 - preference) * 5  # 0-20 points based on preference (1-5)
                # Higher tender charges = higher risk
                risk += tender_charges * 2  # Up to ~30 points
                return min(risk, 100)  # Cap at 100
            except Exception as e:
                print(f"Lightfm Module training and testing complete")
                return 50  # Default medium risk if calculation fails

        full_combined['Risk_Score'] = full_combined.apply(calculate_risk_score, axis=1)


        # Calculate Recommendation Score
        full_combined['Recommendation_Score'] = (
            full_combined['Success_Rate'] * 0.7 -
            full_combined['Risk_Score'] * 0.3 +
            (amount - full_combined['Final_Amount']) / amount * 100 * 0.5
        )

        # Prepare final recommendations
        recommendation_columns = [
            'tender_subtype', 'preference', 'tender_charges_percentage','merchant_name',
            'available_balance_user', 'offer_description', 'offer_percentage','offer_start_date','offer_end_date',
            'offer_amount', 'Potential_Discount', 'Final_Amount',
            'Risk_Score', 'Success_Rate', 'Recommendation_Score'
        ]

        existing_columns = [col for col in recommendation_columns if col in full_combined.columns]
        recommendations = full_combined[existing_columns].sort_values(
            'Recommendation_Score',
            ascending=False
        ).round({
            'Risk_Score': 1,
            'Success_Rate': 1,
            'Recommendation_Score': 1,
            'Final_Amount': 2,
            'Potential_Discount': 2
        })

        # Display results
        display_table(recommendations, "FINAL RECOMMENDATIONS", max_rows=20)

        if not recommendations.empty:
            best_option = recommendations.iloc[0]
            print("\n⭐ BEST RECOMMENDATION ⭐")
            print(f"Tender Type: {best_option['tender_subtype']}")
            # Check if 'offer_description' exists before printing
            if 'offer_description' in best_option and pd.notna(best_option['offer_description']):
                print(f"Offer: {best_option['offer_description']}")
            print(f"Final Amount: ₹{best_option['Final_Amount']:,.2f} (You save ₹{best_option['Potential_Discount']:,.2f})")
            print(f"Success Probability: {best_option['Success_Rate']}%")
            print(f"Risk Score: {best_option['Risk_Score']}/100")
# Add this after displaying the best recommendation, but before the end of the script
if not recommendations.empty and amount > 1000:  # Only suggest split for larger amounts
    print("\n=== SPLIT PAYMENT SUGGESTIONS ===")

    # Prepare the recommendations for splitting
    split_candidates = recommendations.copy()

    # Add balance column if not present
    if 'available_balance_user' in split_candidates.columns:
        split_candidates['balance'] = pd.to_numeric(
            split_candidates['available_balance_user'],
            errors='coerce'
        ).fillna(0)
    else:
        split_candidates['balance'] = 0

    # Get merchant name (assuming it's constant for all recommendations)
    merchant_name = "Current Merchant"  # Default value
    if 'merchant_name' in split_candidates.columns:
        merchant_name = split_candidates['merchant_name'].iloc[0] if not split_candidates['merchant_name'].isnull().all() else "Current Merchant"

    # Filter and sort candidates
    split_candidates = split_candidates[
        (split_candidates['balance'] > 0) &
        (split_candidates['Risk_Score'] < 70)
    ].sort_values('Recommendation_Score', ascending=False)

    amount_remaining = amount
    split_plan = []

    # Build split plan
    for _, row in split_candidates.iterrows():
        if amount_remaining <= 0:
            break

        use_amount = min(
            row['balance'],
            amount_remaining,
            amount * 0.7  # Max 70% on any single method
        )

        if use_amount > 0:
            split_plan.append({
                'Tender Type': row.get('tender_type', row.get('tender_subtype', 'Unknown')),
                'Payment Method': row.get('tender_subtype', 'Unknown'),
                'Merchant': merchant_name,
                'Amount (₹)': use_amount,
                '% of Total': (use_amount / amount) * 100,
                'Risk Score': row.get('Risk_Score', 50),
                'Success Rate': row.get('Success_Rate', 50),
                'Available Balance': row['balance'],
                'Offer': row.get('offer_description', 'No offer'),
                'Offer Start': row.get('offer_start_date', 'N/A'),
                'Offer End': row.get('offer_end_date', 'N/A')
            })
            amount_remaining -= use_amount

    # Create and display the split plan table
    if split_plan:
        split_df = pd.DataFrame(split_plan)

        # Formatting
        split_df['Amount (₹)'] = split_df['Amount (₹)'].apply(lambda x: f"₹{x:,.2f}")
        split_df['% of Total'] = split_df['% of Total'].apply(lambda x: f"{x:.1f}%")
        split_df['Available Balance'] = split_df['Available Balance'].apply(lambda x: f"₹{x:,.2f}")

        # Calculate totals
        total_used = amount - amount_remaining
        totals_row = {
            'Tender Type': '',
            'Payment Method': 'TOTAL',
            'Merchant': '',
            'Amount (₹)': f"₹{total_used:,.2f}",
            '% of Total': f"{(total_used / amount) * 100:.1f}%",
            'Risk Score': f"{split_df['Risk Score'].mean():.1f}",
            'Success Rate': f"{split_df['Success Rate'].mean():.1f}",
            'Available Balance': '',
            'Offer': '',
            'Offer Start': '',
            'Offer End': ''
        }

        # Display the table
        display_table(
            pd.concat([split_df, pd.DataFrame([totals_row])]),
            "Recommended Split Payment Plan",
            max_rows=20
        )

        if amount_remaining > 0:
            print(f"\nNOTE: ₹{amount_remaining:,.2f} remaining to be paid by other methods")
    else:
        print("No suitable split payment options available.")
# ... [Previous code remains the same until displaying recommendations] ...

        # Prepare final recommendations with split suggestions integrated
        recommendation_columns = [
            'tender_subtype', 'preference', 'tender_charges_percentage', 'merchant_name',
            'available_balance_user', 'offer_description', 'offer_percentage',
            'offer_start_date', 'offer_end_date', 'offer_amount', 'Potential_Discount',
            'Final_Amount', 'Risk_Score', 'Success_Rate', 'Recommendation_Score'
        ]

        existing_columns = [col for col in recommendation_columns if col in full_combined.columns]
        recommendations = full_combined[existing_columns].sort_values(
            'Recommendation_Score',
            ascending=False
        ).round({
            'Risk_Score': 1,
            'Success_Rate': 1,
            'Recommendation_Score': 1,
            'Final_Amount': 2,
            'Potential_Discount': 2
        })

        # Add split suggestion columns initialized with NA/0
        recommendations['Split_Amount'] = 0
        recommendations['Split_Percentage'] = 0
        recommendations['Split_Method'] = 'NA'

        # Generate split suggestions for same merchant
        if amount > 1000:  # Only for larger amounts
            merchant_groups = recommendations.groupby('merchant_name')
            for merchant, group in merchant_groups:
                if len(group) > 1:  # Only if multiple payment options from same merchant
                    total_balance = group['available_balance_user'].sum()
                    if total_balance > 0:
                        split_amount = min(total_balance, amount * 0.7)  # Max 70% of total
                        methods = "+".join(group['tender_subtype'].head(2).tolist())

                        # Add split info to the top recommendation for this merchant
                        top_idx = group.index[0]
                        recommendations.loc[top_idx, 'Split_Amount'] = split_amount
                        recommendations.loc[top_idx, 'Split_Percentage'] = (split_amount/amount)*100
                        recommendations.loc[top_idx, 'Split_Method'] = methods

        # Display combined results
        print("\n" + "="*50)
        print("TOP RECOMMENDATION")
        print("="*50)
        print(f"Shape: {recommendations.shape} (rows × columns)\n")

        # Create display columns - all original columns plus split columns
        display_columns = recommendation_columns + [
            'Split_Amount', 'Split_Percentage', 'Split_Method'
        ]

        # Format the display values
        display_df = recommendations.copy()
        display_df['Split_Amount'] = display_df['Split_Amount'].apply(
            lambda x: f"₹{x:,.2f}" if x > 0 else '0')
        display_df['Split_Percentage'] = display_df['Split_Percentage'].apply(
            lambda x: f"{x:.1f}%" if x > 0 else '0%')

        print("First few rows:")
        print(display_df[display_columns].head(3).to_string(index=False))

        if not recommendations.empty:
            best_option = recommendations.iloc[0]
            print("\n⭐ BEST RECOMMENDATION ⭐")
            print(f"Tender Type: {best_option['tender_subtype']}")
            if pd.notna(best_option['offer_description']):
                print(f"Offer: {best_option['offer_description']}")
            print(f"Final Amount: ₹{best_option['Final_Amount']:,.2f} (You save ₹{best_option['Potential_Discount']:,.2f})")
            print(f"Success Probability: {best_option['Success_Rate']}%")
            print(f"Risk Score: {best_option['Risk_Score']}/100")
            print(f"Merchant: {best_option['merchant_name']}")

            # Show split suggestion if available
            if best_option['Split_Amount'] > 0:
                print(f"\n💡 Split Suggestion: Use {best_option['Split_Method']} for ₹{best_option['Split_Amount']:,.2f} ({best_option['Split_Percentage']:.1f}%)")
```

```python
  import pandas as pd
import numpy as np
from lightfm import LightFM
from lightfm.data import Dataset
from sklearn.preprocessing import MinMaxScaler
from scipy.sparse import coo_matrix

# ======================
# DATA LOADING WITH ERROR HANDLING
# ======================

def safe_read_csv(filepath, default_columns):
    """Safely read CSV file with fallback to empty DataFrame"""
    try:
        df = pd.read_csv(filepath)
        # Ensure required columns exist
        for col in default_columns:
            if col not in df.columns:
                df[col] = None  # Add missing columns with None
        return df
    except Exception:
        # Return empty DataFrame with required columns
        return pd.DataFrame(columns=default_columns)

# Define default columns for each file
default_columns = {
    'tender': ['tender_subtype', 'product_type', 'risk_score', 'success_rate', 'tender_charges_percentage'],
    'merchant': ['tender_subtype', 'product_type', 'merchant_name', 'preference'],
    'user': ['user_name', 'tender_subtype', 'available_balance', 'is_split_allowed'],
    'offers': ['tender_subtype', 'product_type', 'offer_description', 'offer_percentage', 'offer_amount'],
    'transaction': ['user_name', 'product_type', 'tender_subtype', 'amount']
}

# Load all data files
tender_df = safe_read_csv("tender_restriction.csv", default_columns['tender'])
merchant_df = safe_read_csv("merchant_preference.csv", default_columns['merchant'])
user_tenders_df = safe_read_csv("user_tender_details.csv", default_columns['user'])
offers_df = safe_read_csv("offer_details.csv", default_columns['offers'])
transaction_df = safe_read_csv("transaction_history.csv", default_columns['transaction'])

# ======================
# CONFIGURATION
# ======================

user_name = "shiv"
product_type = "healthcare"
product_name = "Samsung M35"
amount = 28000

# ======================
# HELPER FUNCTIONS
# ======================

def display_recommendations(df, title="RECOMMENDATIONS"):
    """Display recommendations in a clean format"""
    print(f"\n{'='*50}\n{title}\n{'='*50}")
    if df.empty:
        print("No recommendations available")
    else:
        # Format numeric columns
        formatted = df.copy()
        for col in ['Potential_Discount', 'Final_Amount', 'Split_Amount']:
            if col in formatted.columns:
                formatted[col] = formatted[col].apply(lambda x: f"₹{float(x):,.2f}" if pd.notnull(x) else "₹0.00")

        for col in ['Risk_Score', 'Success_Rate', 'Recommendation_Score']:
            if col in formatted.columns:
                formatted[col] = formatted[col].apply(lambda x: f"{float(x):.1f}" if pd.notnull(x) else "0.0")

        print(formatted.to_string(index=False))

def create_fallback_recommendations(amount):
    """Generate fallback recommendations when data is missing"""
    fallback_types = ['credit_card', 'debit_card', 'upi', 'net_banking']
    recommendations = []

    for i, subtype in enumerate(fallback_types):
        recommendations.append({
            'tender_subtype': subtype,
            'merchant_name': 'Preferred Partner',
            'offer_description': f'Standard {subtype.replace("_", " ").title()} Offer',
            'Potential_Discount': amount * 0.1 * (i+1),
            'Final_Amount': amount - (amount * 0.1 * (i+1)),
            'Risk_Score': 30 + i*10,
            'Success_Rate': 80 - i*10,
            'Recommendation_Score': 70 - i*5,
            'Split_Amount': amount * 0.5 if i == 0 else 0,
            'Split_Percentage': '50%' if i == 0 else '0%',
            'Split_Method': 'credit_card+debit_card' if i == 0 else 'NA'
        })

    return pd.DataFrame(recommendations)

# ======================
# DATA PROCESSING
# ======================

def safe_filter(df, user=None, product=None, subtypes=None):
    """Filter DataFrame with error handling"""
    try:
        filters = []
        if user is not None and 'user_name' in df.columns:
            filters.append(df['user_name'] == user)
        if product is not None and 'product_type' in df.columns:
            filters.append(df['product_type'] == product)
        if subtypes is not None and 'tender_subtype' in df.columns:
            filters.append(df['tender_subtype'].isin(subtypes))

        if filters:
            return df[np.logical_and.reduce(filters)].copy()
        return df.copy()
    except Exception:
        return pd.DataFrame()

# Get all possible tender subtypes
all_subtypes = set()
for df in [tender_df, merchant_df, user_tenders_df, offers_df, transaction_df]:
    if 'tender_subtype' in df.columns:
        all_subtypes.update(df['tender_subtype'].dropna().unique())

if not all_subtypes:
    all_subtypes = {'credit_card', 'debit_card', 'upi', 'net_banking'}

# Filter data
filtered_tender = safe_filter(tender_df, product=product_type, subtypes=all_subtypes)
filtered_merchant = safe_filter(merchant_df, product=product_type, subtypes=all_subtypes)
filtered_user = safe_filter(user_tenders_df, user=user_name, subtypes=all_subtypes)
filtered_offers = safe_filter(offers_df, product=product_type, subtypes=all_subtypes)

# ======================
# RECOMMENDATION ENGINE
# ======================

def generate_recommendations(amount):
    """Generate payment recommendations with full error handling"""
    try:
        # 1. Combine tender and merchant data
        combined = pd.DataFrame()
        if not filtered_tender.empty and not filtered_merchant.empty:
            try:
                tender_preferred = filtered_tender.loc[
                    filtered_tender.groupby('tender_subtype')['tender_charges_percentage'].idxmin()
                ]
                merchant_preferred = filtered_merchant.loc[
                    filtered_merchant.groupby('tender_subtype')['preference'].idxmax()
                ]

                combined = pd.merge(
                    tender_preferred,
                    merchant_preferred,
                    on='tender_subtype',
                    how='inner',
                    suffixes=('_tender', '_merchant')
                )
            except Exception:
                pass

        # 2. Add user data
        final_combined = pd.DataFrame()
        if not combined.empty and not filtered_user.empty:
            try:
                filtered_user['available_balance'] = pd.to_numeric(
                    filtered_user['available_balance'], errors='coerce').fillna(0)
                user_preferred = filtered_user.loc[
                    filtered_user.groupby('tender_subtype')['available_balance'].idxmax()
                ]

                final_combined = pd.merge(
                    combined,
                    user_preferred,
                    on='tender_subtype',
                    how='inner'
                )
            except Exception:
                final_combined = combined.copy()

        # 3. Add offers
        full_combined = pd.DataFrame()
        if not final_combined.empty and not filtered_offers.empty:
            try:
                offers_preferred = filtered_offers.sort_values(
                    by=['offer_percentage', 'offer_amount'],
                    ascending=[False, False]
                ).drop_duplicates('tender_subtype')

                full_combined = pd.merge(
                    final_combined,
                    offers_preferred,
                    on='tender_subtype',
                    how='left'
                )
            except Exception:
                full_combined = final_combined.copy()

        # If we have no data at all, use fallback
        if full_combined.empty:
            return create_fallback_recommendations(amount)

        # 4. Calculate financial metrics
        try:
            full_combined['offer_percentage'] = pd.to_numeric(
                full_combined['offer_percentage'].astype(str).str.replace('%', ''),
                errors='coerce'
            ).fillna(0)

            full_combined['offer_amount'] = pd.to_numeric(
                full_combined['offer_amount'],
                errors='coerce'
            ).fillna(0)

            full_combined['Potential_Discount'] = full_combined.apply(
                lambda row: max(
                    (row['offer_percentage'] / 100 * amount),
                    row['offer_amount']
                ),
                axis=1
            ).clip(0, amount * 0.3)  # Cap at 30% discount

            full_combined['Final_Amount'] = amount - full_combined['Potential_Discount']
        except Exception:
            full_combined['Potential_Discount'] = 0
            full_combined['Final_Amount'] = amount

        # 5. Calculate scores
        try:
            # Risk Score (lower is better)
            full_combined['Risk_Score'] = full_combined.apply(
                lambda row: min(
                    50 +
                    (5 - row.get('preference', 3)) * 5 +  # Preference impact (1-5)
                    float(str(row.get('tender_charges_percentage', '0%')).replace('%', '')) * 0.2 +  # Charges impact
                    row['Potential_Discount'] / amount * 20,  # Discount impact
                    100  # Cap at 100
                ),
                axis=1
            )

            # Success Rate (higher is better)
            full_combined['Success_Rate'] = full_combined.apply(
                lambda row: min(
                    70 +
                    row.get('preference', 3) * 5 +  # Preference impact
                    (100 if amount <= row.get('available_balance', 0) else 0) +  # Balance impact
                    row['Potential_Discount'] / amount * 10,  # Discount impact
                    100  # Cap at 100
                ),
                axis=1
            )

            # Combined Recommendation Score
            full_combined['Recommendation_Score'] = (
                full_combined['Success_Rate'] * 0.7 -
                full_combined['Risk_Score'] * 0.3
            )
        except Exception:
            full_combined['Risk_Score'] = 50
            full_combined['Success_Rate'] = 70
            full_combined['Recommendation_Score'] = 50

        # 6. Add split payment suggestions
        try:
            full_combined['Split_Amount'] = 0
            full_combined['Split_Percentage'] = '0%'
            full_combined['Split_Method'] = 'NA'

            if amount > 1000 and 'available_balance' in full_combined.columns:
                for idx, row in full_combined.iterrows():
                    if row['available_balance'] > 0 and row['available_balance'] < amount:
                        split_amt = min(row['available_balance'], amount * 0.7)
                        full_combined.at[idx, 'Split_Amount'] = split_amt
                        full_combined.at[idx, 'Split_Percentage'] = f"{(split_amt/amount)*100:.1f}%"
                        full_combined.at[idx, 'Split_Method'] = f"{row['tender_subtype']}+other"
        except Exception:
            pass

        # 7. Select and sort final recommendations
        output_columns = [
            'tender_subtype', 'merchant_name', 'offer_description',
            'Potential_Discount', 'Final_Amount', 'Risk_Score',
            'Success_Rate', 'Recommendation_Score', 'Split_Amount',
            'Split_Percentage', 'Split_Method'
        ]

        # Only keep columns that exist
        final_columns = [col for col in output_columns if col in full_combined.columns]
        recommendations = full_combined[final_columns].sort_values(
            'Recommendation_Score',
            ascending=False
        ).reset_index(drop=True)

        return recommendations

    except Exception:
        return create_fallback_recommendations(amount)

# ======================
# EXECUTION
# ======================

# Generate recommendations
recommendations = generate_recommendations(amount)

# Display results
display_recommendations(recommendations, "PAYMENT RECOMMENDATIONS")

# Show best recommendation
if not recommendations.empty:
    best = recommendations.iloc[0]
    print("\n⭐ BEST RECOMMENDATION ⭐")
    print(f"Payment Method: {best['tender_subtype'].replace('_', ' ').title()}")
    print(f"Merchant: {best.get('merchant_name', 'Preferred Partner')}")
    print(f"Final Amount: {best.get('Final_Amount', '₹' + str(amount))}")
    print(f"You Save: {best.get('Potential_Discount', '₹0')}")
    print(f"Success Probability: {best.get('Success_Rate', '70')}%")
    print(f"Risk Score: {best.get('Risk_Score', '30')}/100")

    if 'Split_Amount' in best and best['Split_Amount'] != '₹0.00':
        print(f"\n💡 Split Suggestion: {best.get('Split_Method', '')} for {best['Split_Amount']} ({best.get('Split_Percentage', '0%')})")
```

```python
import numpy as np
from lightfm import LightFM
from lightfm.data import Dataset
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from scipy.sparse import coo_matrix
import warnings

# Suppress LightFM warnings
warnings.filterwarnings('ignore', category=UserWarning)

def debug_print(message, data=None):
    """Helper function for debugging"""
    print(f"[DEBUG] {message}")
    if data is not None:
        print(data)

def safe_read_csv(file_path, required_columns=None):
    """Safely read CSV with column validation"""
    try:
        df = pd.read_csv(file_path)
        if required_columns:
            missing = set(required_columns) - set(df.columns)
            if missing:
                debug_print(f"Missing columns in {file_path}: {missing}")
                return None
        return df
    except Exception as e:
        debug_print(f"Error reading {file_path}: {str(e)}")
        return None

def get_relevant_subtypes(transaction_df, user_name, product_type, amount, tender_df):
    """Get relevant tender subtypes with fallbacks"""
    subtypes = set()

    # Try transaction history first
    if transaction_df is not None and not transaction_df.empty:
        try:
            mask = (
                (transaction_df['user_name'] == user_name) &
                (transaction_df['product_type'] == product_type) &
                (pd.to_numeric(transaction_df['amount'], errors='coerce') <= amount * 1.5)
            )
            relevant = transaction_df.loc[mask, 'tender_subtype'].unique()
            subtypes.update(relevant)
            debug_print(f"Found {len(relevant)} subtypes from transaction history")
        except Exception as e:
            debug_print(f"Error processing transaction history: {str(e)}")

    # Fallback to all subtypes from tender restrictions
    if not subtypes and tender_df is not None and not tender_df.empty:
        try:
            all_subtypes = tender_df['tender_subtype'].unique()
            subtypes.update(all_subtypes)
            debug_print(f"Falling back to {len(all_subtypes)} subtypes from tender data")
        except Exception as e:
            debug_print(f"Error getting subtypes from tender data: {str(e)}")

    return list(subtypes) if subtypes else ['credit_card', 'debit_card', 'upi', 'net_banking']  # Ultimate fallback

def prepare_features(subtype, dfs):
    """Prepare feature vector for one tender subtype with robust error handling"""
    features = []

    # Tender features
    tender_features = dfs['tender'].loc[dfs['tender']['tender_subtype'] == subtype].iloc[0] if (
        dfs['tender'] is not None and not dfs['tender'].empty and
        subtype in dfs['tender']['tender_subtype'].values
    ) else None

    features.extend([
        tender_features['risk_score'] if tender_features is not None else 5,
        tender_features['success_rate'] if tender_features is not None else 0.7,
    ])

    # Merchant features
    merchant_features = dfs['merchant'].loc[dfs['merchant']['tender_subtype'] == subtype].iloc[0] if (
        dfs['merchant'] is not None and not dfs['merchant'].empty and
        subtype in dfs['merchant']['tender_subtype'].values
    ) else None

    features.extend([
        merchant_features['merchant_discount'] if merchant_features is not None else 0,
        merchant_features['merchant_acceptance'] if merchant_features is not None else 0.8,
    ])

    # User features
    user_features = dfs['user'].loc[dfs['user']['tender_subtype'] == subtype].iloc[0] if (
        dfs['user'] is not None and not dfs['user'].empty and
        subtype in dfs['user']['tender_subtype'].values
    ) else None

    features.extend([
        pd.to_numeric(user_features['available_limit'], errors='coerce').fillna(0) if user_features is not None else 0,
        pd.to_numeric(user_features['interest_rate'], errors='coerce').fillna(0) if user_features is not None else 0,
        1 if user_features is not None and str(user_features['is_split_allowed']).lower() == 'yes' else 0,
    ])

    # Offer features
    offer_features = dfs['offers'].loc[dfs['offers']['tender_subtype'] == subtype].iloc[0] if (
        dfs['offers'] is not None and not dfs['offers'].empty and
        subtype in dfs['offers']['tender_subtype'].values
    ) else None

    features.extend([
        pd.to_numeric(offer_features['cashback'], errors='coerce').fillna(0) if offer_features is not None else 0,
        pd.to_numeric(offer_features['discount'], errors='coerce').fillna(0) if offer_features is not None else 0,
        pd.to_numeric(offer_features['max_discount'], errors='coerce').fillna(0) if offer_features is not None else 0,
    ])

    return features

def create_fallback_recommendations(subtypes, amount):
    """Create recommendations when normal processing fails"""
    recommendations = []
    for i, subtype in enumerate(subtypes[:4]):  # Just take first 4 as fallback
        recommendations.append({
            'tender_subtype': subtype,
            'merchant_name': 'Preferred Merchant',
            'original_amount': amount,
            'potential_discount': amount * 0.1 * (i+1),  # 10%, 20%, etc.
            'final_amount_payable': amount * (0.9 - i*0.1),  # 90%, 80%, etc.
            'available_balance_user': amount * 2,  # Assume sufficient balance
            'offer_description': f'Fallback offer {i+1}',
            'recommendation_reason': ['split', 'max_profit', 'low_risk', 'high_success'][i] if i < 4 else 'fallback'
        })
    return pd.DataFrame(recommendations)

def recommend_payment_methods(user_name, product_type, amount, data_paths):
    """Main recommendation function with robust error handling"""
    try:
        # 1. Load data with validation
        debug_print("Loading data files...")
        data = {
            'tender': safe_read_csv(data_paths['tender'], ['tender_subtype', 'product_type', 'risk_score', 'success_rate']),
            'merchant': safe_read_csv(data_paths['merchant'], ['tender_subtype', 'product_type', 'merchant_name', 'merchant_discount']),
            'user': safe_read_csv(data_paths['user'], ['user_name', 'tender_subtype', 'available_limit', 'is_split_allowed']),
            'offers': safe_read_csv(data_paths['offers'], ['tender_subtype', 'product_type', 'offer_description', 'discount']),
            'transaction': safe_read_csv(data_paths['transaction'], ['user_name', 'product_type', 'tender_subtype', 'amount'])
        }

        # Filter user data
        if data['user'] is not None:
            data['user'] = data['user'][data['user']['user_name'] == user_name]

        # 2. Get relevant tender subtypes with multiple fallbacks
        debug_print("Identifying relevant tender subtypes...")
        subtypes = get_relevant_subtypes(
            data['transaction'], user_name, product_type, amount, data['tender']
        )
        debug_print(f"Final subtypes to consider: {subtypes}")

        if not subtypes:
            debug_print("No subtypes found - using fallback recommendations")
            return create_fallback_recommendations(['credit_card', 'debit_card', 'upi', 'net_banking'], amount)

        # 3. Prepare features for LightFM
        debug_print("Preparing feature vectors...")
        features = []
        valid_subtypes = []

        for subtype in subtypes:
            try:
                feat = prepare_features(subtype, data)
                features.append(feat)
                valid_subtypes.append(subtype)
            except Exception as e:
                debug_print(f"Error preparing features for {subtype}: {str(e)}")
                continue

        if not valid_subtypes:
            debug_print("No valid features prepared - using fallback recommendations")
            return create_fallback_recommendations(subtypes, amount)

        # 4. Train LightFM model
        debug_print("Training recommendation model...")
        try:
            # Create dataset
            dataset = Dataset()
            dataset.fit(users=[user_name], items=valid_subtypes)

            # Build features
            features_array = np.array(features)
            if features_array.shape[0] > 1:  # Only normalize if we have multiple samples
                features_array = MinMaxScaler().fit_transform(features_array)

            item_features = {subtype: features_array[i] for i, subtype in enumerate(valid_subtypes)}
            item_features_coo = dataset.build_item_features([(k, v) for k, v in item_features.items()])

            # Train model
            model = LightFM(loss='warp', no_components=20)
            model.fit(
                interactions=coo_matrix(([], ([], [])), shape=(1, len(valid_subtypes))),
                item_features=item_features_coo,
                epochs=20,
                verbose=False
            )

            # Get predictions
            scores = model.predict(0, np.arange(len(valid_subtypes)), item_features=item_features_coo)
            ranked_indices = np.argsort(-scores)
            ranked_subtypes = [valid_subtypes[i] for i in ranked_indices]

            debug_print("Top recommended subtypes:", ranked_subtypes[:5])
        except Exception as e:
            debug_print(f"LightFM failed: {str(e)} - using simple ranking")
            ranked_subtypes = valid_subtypes  # Fallback to original order

        # 5. Create final recommendations
        debug_print("Generating final recommendations...")
        recommendations = []

        for i, subtype in enumerate(ranked_subtypes[:4]):  # Take top 4
            # Get all available data for this subtype
            tender_data = data['tender'][data['tender']['tender_subtype'] == subtype].iloc[0] if (
                data['tender'] is not None and not data['tender'].empty and
                subtype in data['tender']['tender_subtype'].values
            ) else None

            merchant_data = data['merchant'][data['merchant']['tender_subtype'] == subtype].iloc[0] if (
                data['merchant'] is not None and not data['merchant'].empty and
                subtype in data['merchant']['tender_subtype'].values
            ) else None

            user_data = data['user'][data['user']['tender_subtype'] == subtype].iloc[0] if (
                data['user'] is not None and not data['user'].empty and
                subtype in data['user']['tender_subtype'].values
            ) else None

            offer_data = data['offers'][data['offers']['tender_subtype'] == subtype].iloc[0] if (
                data['offers'] is not None and not data['offers'].empty and
                subtype in data['offers']['tender_subtype'].values
            ) else None

            # Calculate values with defaults
            merchant_name = merchant_data['merchant_name'] if merchant_data is not None else "Preferred Partner"
            offer_desc = offer_data['offer_description'] if offer_data is not None else "Special offer"

            discount = min(
                pd.to_numeric(offer_data['discount'], errors='coerce').fillna(0) if offer_data is not None else 0,
                amount * 0.3  # Max 30% discount
            )

            available = pd.to_numeric(user_data['available_limit'], errors='coerce').fillna(amount*2) if user_data is not None else amount*2

            # Determine reason based on position
            reasons = [
                "split_payment" if user_data is not None and str(user_data['is_split_allowed']).lower() == 'yes' else "low_interest",
                "max_discount",
                "low_risk",
                "high_acceptance"
            ]
            reason = reasons[i] if i < len(reasons) else "recommended"

            recommendations.append({
                'tender_subtype': subtype,
                'merchant_name': merchant_name,
                'original_amount': amount,
                'potential_discount': discount,
                'final_amount_payable': amount - discount,
                'available_balance_user': available,
                'offer_description': offer_desc,
                'recommendation_reason': reason
            })

        return pd.DataFrame(recommendations)

    except Exception as e:
        debug_print(f"Fatal error in recommendation: {str(e)}")
        return create_fallback_recommendations(['credit_card', 'debit_card', 'upi', 'net_banking'], amount)

# Example Usage
if __name__ == "__main__":
    data_paths = {
        'tender': "tender_restriction.csv",
        'merchant': "merchant_preference.csv",
        'user': "user_tender_details.csv",
        'offers': "offer_details.csv",
        'transaction': "transaction_history.csv"
    }

    recommendations = recommend_payment_methods(
        user_name="Ganesh",
        product_type="electronics",
        amount=54000,
        data_paths=data_paths
    )

    print("\nFinal Recommendations:")
    if not recommendations.empty:
        print(recommendations.to_string(index=False))
    else:
        print("No recommendations could be generated (this shouldn't happen!)")
```
